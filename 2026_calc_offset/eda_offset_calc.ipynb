```python
import pandas as pd

# Define data path
data_path = '/Users/Mach/dev/aps/data/2026_Dmodel_data'
```

```python
# Read car parquet file
car_df = pd.read_parquet(f'{data_path}/master_dataset_car.parquet')
print(f'Car dataframe shape: {car_df.shape}')
print(f'Car dataframe columns: {car_df.columns.tolist()}')
car_df.head()
```

```python
# Read VIN_BSST parquet file
vin_bsst_df = pd.read_parquet(f'{data_path}/VIN_BSST.parquet')
print(f'VIN_BSST dataframe shape: {vin_bsst_df.shape}')
print(f'VIN_BSST dataframe columns: {vin_bsst_df.columns.tolist()}')
vin_bsst_df.head()
```

```python
# Get unique VINs from car dataframe
car_vins = set(car_df['vin'].unique())
print(f'Total unique VINs in car data: {len(car_vins)}')

# Get unique VINs from BSST dataframe
# Check the column name for VIN in BSST
print(f'VIN_BSST columns: {vin_bsst_df.columns.tolist()}')
```

```python
# Find VINs without BSST
# Assuming VIN column in BSST is 'VIN' or 'vin'
vin_col = 'VIN' if 'VIN' in vin_bsst_df.columns else 'vin'
bsst_vins = set(vin_bsst_df[vin_col].unique())
print(f'Total unique VINs in BSST data: {len(bsst_vins)}')

# VINs in car data that don't have BSST
vins_without_bsst = car_vins - bsst_vins
print(f'\n=== RESULT ===' )
print(f'Number of VINs without BSST: {len(vins_without_bsst)}')
print(f'Percentage of VINs without BSST: {len(vins_without_bsst)/len(car_vins)*100:.2f}%')
```
